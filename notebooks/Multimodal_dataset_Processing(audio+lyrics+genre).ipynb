````python
import pandas as pd

# Load the CSV file into a pandas DataFrame
audio_df = pd.read_csv('/content/drive/MyDrive/MERGE_Audio_Complete/merge_audio_complete_metadata.csv')

# Display the first 5 rows of the DataFrame
display(audio_df.head())
````

````python
import pandas as pd

# Load the new CSV file into a pandas DataFrame
df_lyrics = pd.read_csv('/content/drive/MyDrive/MERGE_Lyrics_Complete/merge_lyrics_complete_metadata.csv')

# Display the first 5 rows of the DataFrame
display(df_lyrics.head())
````

````python
audio_df.rename(columns={'Song': 'song_audio'}, inplace=True)
df_lyrics.rename(columns={'Song': 'song_lyrics'}, inplace=True)

# Display the first 5 rows of the updated DataFrames to confirm the renaming
display(audio_df.head())
display(df_lyrics.head())
````

````python
# Merge the two DataFrames on 'Artist' and 'Title'
merged_df = pd.merge(audio_df, df_lyrics, on=['Artist', 'Title'], how='inner', suffixes=('_audio', '_lyrics'))

# Select the desired columns from the merged DataFrame
final_df = merged_df[[
    'song_audio',
    'song_lyrics',
    'Quadrant_audio',
    'AllMusic Id_audio',
    'Artist',
    'Title',
    'Genres_audio'# Assuming 'Styles' from audio_df is what the user refers to as 'Genres'
]].copy()

# Rename columns to match the user's request
final_df.rename(columns={
    'Quadrant_audio': 'Quadrant',
    'AllMusic Id_audio': 'AllMusic Id',
    'Genres_audio': 'Genres'
}, inplace=True)

# Display the first 5 rows of the combined dataset
display(final_df.head())
final_df.shape
````

````python
import os
count = 0
def get_lyrics_from_file(row):
    global count
    count += 1
    quadrant = row['Quadrant']
    song_lyrics_id = row['song_lyrics']
    file_path = f"/content/drive/MyDrive/MERGE_Lyrics_Complete/{quadrant}/{song_lyrics_id}.txt"
    print(f"{count} instances processed")
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        return None  # Or an empty string, or some other indicator
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
        return None

# Apply the function to create the new 'lyrics' column
final_df['lyrics'] = final_df.apply(get_lyrics_from_file, axis=1)

# Display the first few rows of the updated DataFrame with the new 'lyrics' column
display(final_df.head(20))
````

````python
final_df.drop_duplicates(subset=['Artist', 'Title'], inplace=True)

# Display the first few rows of the updated DataFrame and its shape
display(final_df.head())
final_df.shape
````

##Define Spectrogram Extraction Function


````python
import librosa
import numpy as np
count = 0
def extract_spectrogram(file_path, sr=22050):
    """
    Loads an audio file, converts it to mono, resamples it, and computes a Mel-spectrogram.

    Args:
        file_path (str): The path to the audio file.
        sr (int): Target sample rate for the audio. Defaults to 22050 Hz.

    Returns:
        np.ndarray or None: The Mel-spectrogram in decibels as a NumPy array if successful,
                            otherwise None.
    """
    global count
    count+=1
    print(f'processed {count} files')
    try:
        # Load the audio file, resample to target sr, and convert to mono
        y, sr = librosa.load(file_path, sr=sr, mono=True)

        # Compute the Mel-spectrogram
        mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

        # Convert the Mel-spectrogram to decibel units
        mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

        return mel_spectrogram_db

    except FileNotFoundError:
        print(f"Error: Audio file not found at {file_path}")
        return None
    except Exception as e:
        print(f"Error processing audio file {file_path}: {e}")
        return None

print("Libraries loaded and extract_spectrogram function defined.")
````

## Extract Spectrogram Features for Each Song

Apply the `extract_spectrogram` function to each row of the `final_df`. Construct the full path to each MP3 file using the 'Quadrant' and 'song_audio' columns (e.g., `/content/drive/MyDrive/MERGE_Audio_Complete/{Quadrant}/{song_audio}.mp3`). Store the extracted spectrogram feature arrays in a new column named 'spectrogram_features' in the `final_df`.


````python
def process_audio_file(row):
    file_path = f"/content/drive/MyDrive/MERGE_Audio_Complete/{row['Quadrant']}/{row['song_audio']}.mp3"
    return extract_spectrogram(file_path)

# Apply the function to create the new 'spectrogram_features' column
final_df['spectrogram_features'] = final_df.apply(process_audio_file, axis=1)

# Display the first few rows of the updated DataFrame with the new 'spectrogram_features' column
display(final_df.head())
````

## Visualize a Sample Spectrogram

````python
import matplotlib.pyplot as plt
import librosa.display

print("Libraries for visualization loaded.")
````

````python
sample_spectrograms = final_df[final_df['spectrogram_features'].notna()].head(2)

if not sample_spectrograms.empty:
    for index, row in sample_spectrograms.iterrows():
        spectrogram = row['spectrogram_features']
        artist = row['Artist']
        title = row['Title']

        plt.figure(figsize=(10, 4))
        librosa.display.specshow(spectrogram, sr=22050, x_axis='time', y_axis='mel')
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'Mel-Spectrogram: {artist} - {title}')
        plt.xlabel('Time (s)')
        plt.ylabel('Mel Frequency (Hz)')
        plt.tight_layout()
        plt.show()
else:
    print("No valid spectrograms found in the DataFrame to display.")
````
