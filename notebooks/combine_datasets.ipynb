```python
import pandas as pd

gtzan = pd.read_csv('/kaggle/input/cse425-datasets/features_30_sec.csv')
bangla = pd.read_csv('/kaggle/input/cse425-datasets/banglabeats_features_30sec.csv')

# Tag sources
gtzan['language'] = 'gtzan'
bangla['language'] = 'banglabeats'

# Combine them
full_dataset = pd.concat([gtzan, bangla], ignore_index=True)
full_dataset.to_csv("combined_mfcc(GTZAN+BanglaBeats).csv", index=False)
```
